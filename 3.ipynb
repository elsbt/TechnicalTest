```python
population = np.random.randint(20,30,1000)
np.random.seed(10)

estimates=[]

for x in range(200):
  sample = np.random.choice(a=population, size=100)
  estimates.append (sample.mean())
```

```python
np.mean(population)
```

```python
pd.DataFrame(estimates).plot(kind="density")
```

```python
import scipy.stats as stats

z_critical =stats.norm.ppf(q=0.9)
```

```python
t_critical =stats.t.ppf(q=0.9, df=24)
```

```python
margin_of_error = z_critical * (np.std(estimates)/np.sqrt(200))
```

```python
np.mean(estimates) - margin_of_error
```
